```python
def show_summary(df) :
    
    print(80 * "+")
    print(f"DIMENSIONS : ({df.shape[0]}, {df.shape[1]})")
    
    print(80 * "+")
    print("COULMNS: \n")
    print(df.columns.values)
    
    print(80 * "+")
    print("DATA INFO: \n")
    print(df.dtypes)
    
    print(80 * "+")
    print("MISSING VALUES: \n")
    print(df.isnull().sum())
    
    print(80 * "+")
    print("NUMBER OF UNIQUE VALUES: \n")
    print(df.nunique())
```

```python
show_summary(world_df)
```

Subset world_df into diferrent dataframe based on energy source

```python
# function to subset based on the energy source
def subset_energy_source(df: pd.DataFrame, source :str) -> pd.DataFrame:
    
    source_col = df.filter(like=source).columns
    new_df = df[['country', 'year', 'iso_code', 'continent', 'status', 'population', 'hdi_rank_2021', 'hdi_2021', 'gdp', 'gdp_per_capita'] + list(source_col)]
    
    return new_df
        
```

```python
coal_df = subset_energy_source(world_df, "coal")
biofuel_df = subset_energy_source(world_df, "biofuel")
gas_df = subset_energy_source(world_df, "gas")
hydro_df = subset_energy_source(world_df, "hydro")
low_carbon_df = subset_energy_source(world_df,  "low_carbon")
nuclear_df = subset_energy_source(world_df, "nuclear")
oil_df = subset_energy_source(world_df, "nuclear")
solar_df = subset_energy_source(world_df, "solar")
wind_df = subset_energy_source(world_df, "wind")
```
